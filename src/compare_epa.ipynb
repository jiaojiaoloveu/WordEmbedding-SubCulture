```python
import csv
label = {}
label_std = {}
with open('../data/GitHub_Aggregated.csv') as fp:
    reader = csv.DictReader(fp)
    for row in reader:
        concept = row['Concept']
        e, p, a = float(row['Evaluation_mean']), float(row['Potency_mean']), float(row['Activity_mean'])
        label[concept] = [round(d,3) for d in [e, p, a]]
        e, p, a = float(row['Evaluation_std']), float(row['Potency_std']), float(row['Activity_std'])
        label_std[concept] = [round(d,3) for d in [e, p, a]]
```
