```python
import csv
label = {}
with open('../data/GitHub_Aggregated.csv') as fp:
    reader = csv.DictReader(fp)
    for row in reader:
        concept = row['Concept']
        e, p, a = float(row['Evaluation_mean']), float(row['Potency_mean']), float(row['Activity_mean'])
        label[concept] = [round(d,3) for d in [e, p, a]]
```

```python
warriner = {}
with open('../data/epa/Ratings_Warriner_et_al_epa') as fp:
    data = json.load(fp)
    for w in data.keys():
        epa = [data[w]['E'], data[w]['P'], data[w]['A']]
        warriner[w] = [round(d, 3) for d in epa]
```

```python
import json
predict = {}
with open('../result/epa_expansion/nn/nn_result_github_all_2') as fp:
    predict = json.load(fp)
    for w in predict.keys():
        predict[w][2] = -predict[w][2]
        predict[w] = [round(d, 3) for d in predict[w]]
```

```python
google = {}
with open('../result/epa_expansion/nn/nn_result_google_all_2') as fp:
    google = json.load(fp)
    for w in google.keys():
        google[w][2] = -google[w][2]
        google[w] = [round(d, 3) for d in google[w]]
```

```python
wlist = wlist & set(google.keys())
wlist = set(label.keys()) & set(warriner.keys()) & set(predict.keys()) & set(google.keys())
```

```python
print(len(label.keys()))
print(len(wlist))
```

```python
wlist_warriner, wlist_label, wlist_pred, wlist_google = list(), list(), list(), list()
for w in wlist:
    wlist_warriner.append(warriner[w])
    wlist_label.append(label[w])
    wlist_pred.append(predict[w])
    wlist_google.append(google[w])
import numpy as np
wwarriner = np.array(wlist_warriner)
wlabel = np.array(wlist_label)
wpred = np.array(wlist_pred)
wgoogle = np.array(wlist_google)
```

```python
mean = np.mean(np.abs(wlabel - wpred), axis=0)
print(mean)
rmse = np.sqrt(np.mean((wlabel - wpred) ** 2, axis=0))
print(rmse)
```

```python
mean = np.mean(np.abs(wlabel - wgoogle), axis=0)
print(mean)
rmse = np.sqrt(np.mean((wlabel - wgoogle) ** 2, axis=0))
print(rmse)
```

```python
mean = np.mean(np.abs(wlabel - wwarriner), axis=0)
print(mean)
rmse = np.sqrt(np.mean((wlabel - wwarriner) ** 2, axis=0))
print(rmse)
```

```python
epa_dic = {}
for w in wlist:
    k = '%s %s %s' % (w, label[w], predict[w])
    epa_dic[k] = abs(label[w][0] - predict[w][0]) + abs(label[w][1] - predict[w][1]) \
    + abs(label[w][2] - predict[w][2])
from collections import OrderedDict
ordered_epa_dic = OrderedDict(sorted(epa_dic.items(), key=lambda x: x[1]))
for key in ordered_epa_dic:
    print(key)
```
