```python
import csv
label = {}
with open('../data/R_Concepts.txt', encoding = "ISO-8859-1") as fp:
    reader = csv.DictReader(fp)
    for row in reader:
        concept = row['Concept']
        e, p, a = float(row['White__E_Mean']), float(row['White__P_Mean']), float(row['White__A_Mean'])
        label[concept] = [e, p, a]
```

```python
import json
predict = {}
with open('../result/epa_expansion/nn/nn_result_github_all') as fp:
    predict = json.load(fp)
```

```python
wlist = set(label.keys()) & set(predict.keys())
```

```python
print(len(label.keys()))
print(len(wlist))
```

```python
wlist_label, wlist_pred = list(), list()
for w in wlist:
    wlist_label.append(label[w])
    wlist_pred.append(predict[w])
print(len(wlist_label))
print(len(wlist_pred))
```

```python
import numpy as np
wlabel = np.array(wlist_label)
wpred = np.array(wlist_pred)
print(wlabel.shape)
print(wpred.shape)
```

```python
wlabel_mask = wlabel > 0
wpred_mask = wpred > 0
mask_sum = np.sum(wlabel_mask & wpred_mask, axis=0)
print(mask_sum)
```

```python
mean = np.mean(np.abs(wlabel - wpred), axis=0)
print(mean)
```
