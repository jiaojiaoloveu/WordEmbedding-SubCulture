```python
%matplotlib inline
import numpy as np
import pandas as pd
import json
import os
from scipy import stats, integrate
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
os.chdir('..')
%pwd
```

```python
with open('../result/state_prediction/epa', 'r') as fp:
    epa = np.array(json.load(fp))
print(epa.shape)
```

```python
def plot(axis):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    surf = ax.plot_trisurf(epa[:,axis,1], epa[:,axis,0], epa[:,axis,2], cmap=plt.cm.viridis, linewidth=0.1)
    fig.colorbar(surf, shrink=0.5, aspect=5)
    plt.show()
```

```python
plot(0)
```

```python
plot(1)
```

```python
plot(2)
```

```python
plot(3)
```

```python
def plot2(axis):
    sns.distplot(epa[:, axis, 0], color='r')
    sns.distplot(epa[:, axis, 1], color='b')
    sns.distplot(epa[:, axis, 2], color='g')
```

```python
print(np.mean(epa, axis=0))
print(np.std(epa, axis=0))
```

```python
plot2(0)
```

```python
epa = epa - np.mean(epa, axis=0)
epa = epa / np.std(epa, axis=0)
```

```python
plot2(0)
```

```python
plot2(1)
```
