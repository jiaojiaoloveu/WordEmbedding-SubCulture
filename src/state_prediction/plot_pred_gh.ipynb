```python
with open('../result/state_prediction/gh_sent_svo', 'r') as fp:
    svo_sentence = json.load(fp)
```

```python
basic_sentiment = ['aggressive', 'angry', 'calm', 'careless',
                   'cautious', 'defensive', 'happy', 'nervous',
                   'sorry', 'thanks'
                   ]
with open('../result/state_prediction/svo_senti_github_comment', 'r') as fp:
    senti = json.load(fp)
for w in senti.keys():
    print(w)
    flag = zip(basic_sentiment, senti[w][1])
    flag_sent = [t[0] for t in flag if t[1] == '1']
    senti[w][1] = flag_sent
```

```python
len(senti.keys())
```

```python
count = 0
for w in senti.keys():
    pred = [senti[w][0][i][0] for i in [0, 1, 2]]
    truth = senti[w][1]
    if len(set(pred) & set(truth)) > 0:
        print(w)
        print(svo_sentence[w])
        print(pred)
        print(truth)
        print('==========')
        count += 1
```
