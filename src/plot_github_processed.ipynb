```python
import pandas as pd
raw_data = pd.read_csv('../data/GitHub_Processed.csv')
raw_data = raw_data[['Participant ID', 'Concept', 'Evaluation', 'Potency', 'Activity']]
print(raw_data)
```

```python
concepts = set(raw_data['Concept'].values.tolist())
print(len(concepts))
print(concepts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
def plot(word):
    pd = raw_data.loc[raw_data['Concept'] == word]
    print(word)
    print('count:')
    print(pd.shape[0])
    print('mean: ')
    print(pd.mean()[1:])
    print('std: ')
    print(pd.std()[1:])
    print(pd)
    haxis = ['Evaluation', 'Potency', 'Activity']
    fig, axn = plt.subplots(1, 3, sharex=True, sharey=True)
    fig.set_size_inches(15, 4)
    for i, ax in enumerate(axn.flat):
        sns.distplot(pd[haxis[i]], ax=ax)
    print('======================================')
```

```python
plot('quiet')
```

```python
import pandas as pd
gc = pd.read_csv('../data/GitHub_GC_Checks.csv')
list(gc)
gc = gc[['Participant_ID', 'Concept', 'Rating_Total', 'Duration', 
         'AvgSkip', 'AvgDir1', 'AvgDir0', 'AvgDir_1', 'Evaluation']]
gc['Duration per concept'] = gc['Duration'] / gc['Rating_Total']
# Average Time Per Concept (Duration/Rating_Total) < 2 seconds 

# AverageSkip  <.35 (0.032)

# AverageDir (Right) >.85

# AverageDir0 (Neutral) >.85

# AverageDir_1 (Left) > .85
```
