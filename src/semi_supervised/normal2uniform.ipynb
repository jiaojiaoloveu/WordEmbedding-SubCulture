```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
```

```python
mu, sigma = 0, 1
x = np.random.normal(mu, sigma, 10000)
```

```python
count, bins, ignored = plt.hist(x, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), 
         linewidth=2, color='r')
plt.show()
```

```python
print(x)
```

```python
y = 0.5 * special.erfc(-x / np.sqrt(2))
print(y)
```

```python
count, bins, ignored = plt.hist(y, 100, density=True)
plt.plot(bins, np.ones(101))
plt.show()
```

```python
z = -np.sqrt(2) * special.erfcinv(2 * y)
```

```python
count, bins, ignored = plt.hist(z, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), 
         linewidth=2, color='r')
plt.show()
```

```python
print(x)
print(z)
```
